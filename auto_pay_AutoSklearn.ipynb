```python
import pandas as pd
import numpy as np
file_path = '/content/drive/MyDrive/df.csv'

def load_csv(file_path):
    # Load the Excel file
    return pd.read_csv(file_path)

df = load_csv(file_path)

df
```

```python
from sklearn.model_selection import train_test_split
X, y = df.drop('Y_DEN_AMT_CAD_inflated', axis=1), df['Y_DEN_AMT_CAD_inflated']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=42)
```

```python
from autosklearn.regression import AutoSklearnRegressor
from pprint import pprint
from sklearn.metrics import r2_score

automl = AutoSklearnRegressor(
    time_left_for_this_task=3600*3, #3hrs
    #per_run_time_limit=30,
    tmp_folder="/content/drive/MyDrive/automl/temp",
    memory_limit=40000
)

automl.fit(X_train, y_train, dataset_name="autopay")
```

```python
import pickle

# Save the model
with open('/content/drive/MyDrive/automl/autosklearn_trained_model.pkl', 'wb') as f:
    pickle.dump(automl, f)
```

```python
y_pred = automl.predict(X_test)


from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

def evaluate_regression_model(y_test, y_pred, X_test):
    """
    Evaluate the performance of a regression model.

    Parameters:
    y_test (array-like): The actual target values.
    y_pred (array-like): The predicted target values from the model.
    X_test (DataFrame): The test set features.

    Returns:
    dict: A dictionary containing MAE, MSE, RMSE, R-squared, and Adjusted R-squared.
    """
    # Calculate the standard metrics
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)

    # Calculate Adjusted R-squared
    n = len(y_test)  # Number of observations
    p = X_test.shape[1]  # Number of predictors
    adjusted_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)

    # Create a dictionary to hold the metrics
    metrics = {
        'Mean Absolute Error (MAE)': mae,
        'Mean Squared Error (MSE)': mse,
        'Root Mean Squared Error (RMSE)': rmse,
        'R-squared': r2,
        'Adjusted R-squared': adjusted_r2
    }

    return metrics

# Example usage:
metrics = evaluate_regression_model(y_test, y_pred, X_test)
print(metrics)
```
